```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "gym_nutrition_data.csv"  # Update this if needed
df = pd.read_csv(file_path)

# Convert 'Date' column to datetime format
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")


# Calculate total weight lifted (Handling NaN values for Rest Days)
df["Total Weight Lifted"] = df["Sets"] * df["Reps"] * df["Weight Lifted (kg)"]
df["Total Weight Lifted"] = df["Total Weight Lifted"].fillna(0)  # Replace NaN for rest days

# Sort by Date
df = df.sort_values(by="Date")

# Plot 1: Total Weight Lifted Over Time
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Total Weight Lifted"], marker='o', linestyle='-', color='b', label="Total Weight Lifted (kg)")
plt.xlabel("Date")
plt.ylabel("Total Weight Lifted (kg)")
plt.title("Strength Progress Over Time")
plt.xticks(rotation=45)  # Rotate x-axis for better readability
plt.legend()
plt.grid(True)
plt.show()

# Plot 2: Calories Intake Over Time
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Calories"], marker='o', linestyle='-', color='g', label="Calories Intake")
plt.xlabel("Date")
plt.ylabel("Calories")
plt.title("Daily Caloric Intake Trend")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()

# Plot 3: Protein Intake Over Time
plt.figure(figsize=(10, 5))
plt.plot(df["Date"], df["Protein (g)"], marker='o', linestyle='-', color='r', label="Protein Intake (g)")
plt.xlabel("Date")
plt.ylabel("Protein (g)")
plt.title("Daily Protein Consumption Trend")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "gym_nutrition_data.csv"  # Update if needed
df = pd.read_csv(file_path)

# Fill missing values
df["Weight Lifted (kg)"] = df["Weight Lifted (kg)"].fillna(0)
df["Sets"] = df["Sets"].fillna(0)
df["Reps"] = df["Reps"].fillna(0)

# Calculate total weight lifted per workout type
df["Total Weight Lifted"] = df["Sets"] * df["Reps"] * df["Weight Lifted (kg)"]
workout_performance = df.groupby("Workout Type")["Total Weight Lifted"].sum()

# Define bar colors
colors_list = ['blue', 'green', 'red', 'yellow', 'purple']

# Create bar chart
plt.figure(figsize=(8,5))
plt.bar(workout_performance.index, workout_performance.values, color=colors_list)

# Labels & Title
plt.xlabel("Workout Type")
plt.ylabel("Total Weight Lifted (kg)")
plt.title("Total Weight Lifted by Workout Type")

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "/content/gym_nutrition_data.csv"  # Update path if needed
df = pd.read_csv(file_path)

# Group data by Workout Type and sum Calories
calories_per_workout = df.groupby("Workout Type")["Calories"].sum()

# Define bar colors
colors_list = ['blue', 'green', 'red', 'yellow', 'purple']

# Create horizontal bar chart
plt.figure(figsize=(8, 5))
plt.barh(calories_per_workout.index, calories_per_workout.values, color=colors_list)

# Labels & Title
plt.xlabel("Total Calories Consumed")
plt.ylabel("Workout Type")
plt.title("Total Calories Consumed by Workout Type")

# Show plot
plt.show()
```
